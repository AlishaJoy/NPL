```python
import pandas as pd
import numpy as np
```

```python
ds = pd.read_csv('/content/sample_data/Amazon_Unlocked_Mobile.csv',engine ="python",error_bad_lines= False)
ds.head()
```

```python
# drop missing values:
ds.dropna(inplace= True)
```

```python
# remove any 'neutral ratings equal to 3
ds = ds[ds['Rating']!=3]
```

```python
# encode 4s and 5s as 1
#encode 1s and 2s as 0

ds['Positively Rated'] = np.where(ds['Rating']>3,1,0)
ds.head(10)
```

```python
#most rating are positive
ds['Positively Rated'].mean()
```

```python
!pip install scikit_learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ds['Reviews'],ds['Positively Rated'],random_state=0)
 #random_state=0 will maintain the random
 # x train = input
 # y train = label for the input
```

### CountVectorizer

```python
li = ["welcome to this session","it starts at at nine","it ends at 4:30"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#fit the countvectorizer to the training data
vect = CountVectorizer().fit(li)
```

```python
vect.vocabulary_
```

```python
vect.transform(li).toarray() # each array is each sentence, each words repeatation
```

```python
vectt = CountVectorizer(ngram_range=(1,2)).fit(x_train)
vectt.vocabulary_
```

```python
x_train_vectorized = vectt.transform(x_train)
x_train_vectorized
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train_vectorized,y_train)
```

```python
predictions = model.predict(vectt.transform(x_test))
```

```python
predictions
```

```python
from sklearn.metrics import roc_auc_score
print('AUC:',roc_auc_score(y_test,predictions))
```
